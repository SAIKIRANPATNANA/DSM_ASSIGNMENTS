

Q1. Characteristics of tuples:
- Tuples are ordered collections of elements, meaning the order in which the elements are defined is maintained.
- Tuples can contain elements of different data types, including other tuples or mutable objects like lists.
- Tuples are immutable, meaning once they are created, their elements cannot be changed or modified.

Q2. Two tuple methods in Python:
The two built-in tuple methods in Python are `count()` and `index()`.

Example of `count()` method:
```python
my_tuple = (1, 2, 3, 4, 1, 2, 1)
count_of_1 = my_tuple.count(1)
print(count_of_1)  # Output: 3
```

Example of `index()` method:
```python
my_tuple = (1, 2, 3, 4, 1, 2, 1)
index_of_2 = my_tuple.index(2)
print(index_of_2)  # Output: 1
```

Reason for having only two in-built methods for tuples:
Tuples have fewer built-in methods compared to lists because tuples are immutable, so methods like `append()`, `extend()`, `insert()`, and `remove()` which modify the tuple are not applicable.

Q3. Collection datatypes in Python that do not allow duplicate items:
Sets and dictionaries do not allow duplicate items. Here's how you can use a set to remove duplicates from the given list:

```python
my_list = [1, 1, 1, 2, 1, 3, 1, 4, 2, 1, 2, 2, 2, 3, 2, 4, 3, 1, 3, 2, 3, 3, 3, 4, 4, 1, 4, 2, 4, 3, 4, 4]
unique_items = set(my_list)
print(list(unique_items))  # Output: [1, 2, 3, 4]
```

Q4. Difference between `union()` and `update()` methods for a set:
- `union()`: Returns a new set containing all the unique elements from both sets.
- `update()`: Updates the set with the union of itself and others.

Example of `union()` method:
```python
set1 = {1, 2, 3}
set2 = {3, 4, 5}
union_set = set1.union(set2)
print(union_set)  # Output: {1, 2, 3, 4, 5}
```

Example of `update()` method:
```python
set1 = {1, 2, 3}
set2 = {3, 4, 5}
set1.update(set2)
print(set1)  # Output: {1, 2, 3, 4, 5}
```

Q5. Dictionary:
A dictionary is an unordered collection of data in a key:value pair format. Each key is unique and maps to a corresponding value.

Example:
```python
my_dict = {'name': 'John', 'age': 30, 'city': 'New York'}
```
In this dictionary, `'name'`, `'age'`, and `'city'` are keys, and `'John'`, `30`, and `'New York'` are their corresponding values.

A dictionary is unordered, meaning there is no guarantee that the order in which you add items will be preserved.

Q6. Yes, we can create nested dictionaries in Python. A nested dictionary is a dictionary inside another dictionary. Each key in a nested dictionary is associated with another dictionary as its value.

Example of a simple one-level nested dictionary:
```python
nested_dict = {'outer_key': {'inner_key': 'value'}}
```
In this example, `'outer_key'` is the key of the outer dictionary, and its corresponding value is another dictionary `{'inner_key': 'value'}`.

Q7. Using `setdefault()` method to create a key named 'topics' and add the given list as its value to the given dictionary:
```python
dict1 = {'language': 'Python', 'course': 'Data Science Masters'}
dict1.setdefault('topics', []).extend(['Python', 'Machine Learning', 'Deep Learning'])
print(dict1)
```
Output:
```
{'language': 'Python', 'course': 'Data Science Masters', 'topics': ['Python', 'Machine Learning', 'Deep Learning']}
```

Q8. View objects in dictionaries:
There are three view objects in dictionaries: `keys()`, `values()`, and `items()`. These view objects provide a dynamic view on the dictionary's keys, values, and key-value pairs, respectively.

Example using the three view objects in Python:
```python
dict1 = {'Sport': 'Cricket', 'Teams': ['India', 'Australia', 'England', 'South Africa', 'Sri Lanka', 'New Zealand']}

# Display keys
print(dict1.keys())

# Display values
print(dict1.values())

# Display key-value pairs
print(dict1.items())
```
Output:
```
dict_keys(['Sport', 'Teams'])
dict_values(['Cricket', ['India', 'Australia', 'England', 'South Africa', 'Sri Lanka', 'New Zealand']])
dict_items([('Sport', 'Cricket'), ('Teams', ['India', 'Australia', 'England', 'South Africa', 'Sri Lanka', 'New Zealand'])])
```

These view objects provide dynamic views into the dictionary's data, so any changes made to the dictionary will be reflected in the view objects.
