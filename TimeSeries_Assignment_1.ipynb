
### Q1. What is a time series, and what are some common applications of time series analysis?

A time series is a sequence of data points collected or recorded at successive equally spaced intervals over time. Time series analysis involves analyzing and interpreting this data to extract meaningful insights or make predictions about future trends. Common applications include:
- Economic forecasting
- Stock market analysis
- Weather forecasting
- Demand forecasting in sales and inventory management
- Health monitoring in medicine
- Energy consumption prediction

### Q2. What are some common time series patterns, and how can they be identified and interpreted?

Common time series patterns include:
- Trend: Long-term upward or downward movement.
- Seasonality: Periodic fluctuations occurring at fixed intervals.
- Cyclical patterns: Fluctuations not necessarily of fixed periods.
- Irregular or random movements: Unpredictable variations.
Identifying these patterns involves visual inspection of the data using plots like line plots, scatter plots, and decomposing the time series into its components.

### Q3. How can time series data be preprocessed before applying analysis techniques?

Time series data preprocessing steps may include:
- Handling missing values or outliers.
- Resampling to ensure uniform time intervals.
- De-trending and de-seasonalizing.
- Normalizing or scaling.
- Feature engineering to create relevant predictors.
- Splitting into training, validation, and test sets.

### Q4. How can time series forecasting be used in business decision-making, and what are some common challenges and limitations?

Time series forecasting helps businesses make informed decisions by predicting future trends and behaviors. For example, it aids in inventory management, resource allocation, and budget planning. Common challenges include handling noisy data, dealing with unexpected events, and selecting appropriate models. Limitations include uncertainty in predictions, reliance on historical data, and the assumption of stationary data.

### Q5. What is ARIMA modelling, and how can it be used to forecast time series data?

ARIMA (AutoRegressive Integrated Moving Average) is a widely used method for time series forecasting. It combines autoregressive (AR), differencing (I), and moving average (MA) components. ARIMA models capture linear dependencies between observations at different time lags. It can be used to forecast future values based on past observations and trends.

### Q6. How do Autocorrelation Function (ACF) and Partial Autocorrelation Function (PACF) plots help in identifying the order of ARIMA models?

ACF and PACF plots show the correlation of a time series with its lagged values. Peaks or spikes in these plots indicate significant correlations. ACF helps identify the order of MA components, while PACF helps identify the order of AR components. The decay in correlations guides the selection of model orders.

### Q7. What are the assumptions of ARIMA models, and how can they be tested for in practice?

ARIMA models assume stationarity, meaning that statistical properties like mean and variance remain constant over time. This can be tested using statistical tests like the Augmented Dickey-Fuller (ADF) test for unit roots. Other assumptions include linearity and homoscedasticity, which can be assessed through visual inspection and diagnostic checks.

### Q8. Suppose you have monthly sales data for a retail store for the past three years. Which type of time series model would you recommend for forecasting future sales, and why?

Given monthly sales data, a seasonal ARIMA (SARIMA) model might be suitable, as it can capture both the seasonality and any underlying trends in the data. Additionally, incorporating exogenous variables such as promotions or holidays could improve forecast accuracy.

### Q9. What are some of the limitations of time series analysis? Provide an example of a scenario where the limitations of time series analysis may be particularly relevant.

Limitations include the assumption of stationarity, which may not hold in dynamic environments. For instance, in volatile markets or during unprecedented events like the COVID-19 pandemic, historical patterns may not accurately predict future behavior.

### Q10. Explain the difference between a stationary and non-stationary time series. How does the stationarity of a time series affect the choice of forecasting model?

A stationary time series has constant statistical properties over time, such as mean and variance, making it easier to model. In contrast, a non-stationary time series exhibits trends, seasonality, or changing variance, making it more challenging to model. Stationarity affects the choice of forecasting model because many methods, such as ARIMA, require the data to be stationary for accurate predictions. If the time series is non-stationary, pre-processing techniques like differencing may be applied to make it stationary before applying forecasting models.
