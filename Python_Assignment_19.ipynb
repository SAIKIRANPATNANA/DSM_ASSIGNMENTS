Q1. **Create a Pandas Series that contains the following data: 4, 8, 15, 16, 23, and 42. Then, print the series.**

```python
import pandas as pd

# Creating a Pandas Series
data = [4, 8, 15, 16, 23, 42]
series = pd.Series(data)

# Printing the series
print(series)
```

Output:
```
0     4
1     8
2    15
3    16
4    23
5    42
dtype: int64
```

| Name   | Age | Gender |
|--------|-----|--------|
| Alice  | 25  | Female |
| Bob    | 30  | Male   |
| Charlie| 35  | Male   |

Then, print the DataFrame.

Q4. **What is ‘DataFrame’ in pandas and how is it different from pandas.Series? Explain with an example.**

- **DataFrame:** DataFrame is a two-dimensional labeled data structure in pandas that can hold data of different types (e.g., numeric, string, boolean). It is similar to a spreadsheet or SQL table, where data is organized in rows and columns. Each column in a DataFrame is a pandas Series.

- **Example:**
  ```python
  import pandas as pd

  # Creating a DataFrame from a dictionary
  data = {'A': [1, 2, 3],
          'B': ['a', 'b', 'c']}
  df = pd.DataFrame(data)

  # Printing the DataFrame
  print(df)
  ```
  Output:
  ```
     A  B
  0  1  a
  1  2  b
  2  3  c
  ```

  In this example, `df` is a DataFrame with two columns labeled 'A' and 'B', where 'A' and 'B' are pandas Series containing integer and string data, respectively.

- **Difference:**
  - A pandas Series is a one-dimensional array-like object that can hold data of a single data type, while a DataFrame is a two-dimensional structure that can hold data of multiple data types organized in rows and columns.
  - A DataFrame can be thought of as a collection of pandas Series, where each column in the DataFrame is a Series.

Q5. **What are some common functions you can use to manipulate data in a Pandas DataFrame? Can you give an example of when you might use one of these functions?**

- **Common functions for DataFrame manipulation:**
  - `head()`: Returns the first n rows of the DataFrame.
  - `tail()`: Returns the last n rows of the DataFrame.
  - `info()`: Provides a concise summary of the DataFrame, including data types and missing values.
  - `describe()`: Generates descriptive statistics for numerical columns.
  - `loc[]` and `iloc[]`: Accesses rows and columns by label and index, respectively.
  - `drop()`: Removes rows or columns from the DataFrame.
  - `groupby()`: Groups data by one or more columns and performs aggregation operations.
  - `merge()`: Combines multiple DataFrames based on common columns or indices.

- **Example:**
  ```python
  # Using describe() function to generate descriptive statistics
  print(df.describe())
  ```

  This function can be used to quickly analyze the numerical data in a DataFrame, providing insights such as mean, standard deviation, minimum, maximum, and quartile values.

Q7. **Create a DataFrame using multiple Series. Explain with an example.**

```python
import pandas as pd

# Creating multiple Series
name_series = pd.Series(['Alice', 'Bob', 'Charlie'])
age_series = pd.Series([25, 30, 35])
gender_series = pd.Series(['Female', 'Male', 'Male'])

# Creating a DataFrame using the Series
df = pd.DataFrame({'Name': name_series, 'Age': age_series, 'Gender': gender_series})

# Printing the DataFrame
print(df)
```

In this example, we create three pandas Series (`name_series`, `age_series`, and `gender_series`) containing data for the 'Name', 'Age', and 'Gender' columns, respectively. Then, we create a DataFrame `df` by passing these Series as values to a dictionary where the keys are the column names. Finally, we print the DataFrame, which consists of three columns ('Name', 'Age', and 'Gender') and three rows of data.
